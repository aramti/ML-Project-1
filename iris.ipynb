```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url = 'http://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'
df = pd.read_csv(url, names=['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'class'])
df.head()
df.describe()
```

```python
data = df.values
x = data[:,0:4]
y = data[:,4]
print(x)
print(y)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25 )
print(y_test)
```

```python
from sklearn.svm import SVC
model_svc = SVC()
model_svc.fit(x_train, y_train)
predict1 = model_svc.predict(x_test)
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,predict1))
```

```python
from sklearn.linear_model import LogisticRegression
model_lr = LogisticRegression()
model_lr.fit(x_train, y_train)
predict2 = model_lr.predict(x_test)
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,predict2))
```

```python
from sklearn.tree import DecisionTreeClassifier
model_tree = DecisionTreeClassifier()
model_tree.fit(x_train, y_train)
predict3 = model_tree.predict(x_test)
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,predict3))
```

```python
from sklearn.metrics import classification_report
print(y_test, predict3)
print(classification_report(y_test, predict3))
```
